```python
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset=pd.read_csv('Salary_Data.csv')
```

```python
X=dataset.iloc[:,:-1].values
```

```python
y=dataset.iloc[:,-1].values
```

```python
# Splitting into training data and testing data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# we create a new variable of any name which is an instance/object of the class Linear Regression
regressor = LinearRegression()
```

```python
# Method we use to train our model is called fit method
regressor.fit(X_train,y_train)
```

```python
# Predicting the testset results
# y_test contains real salaries and y_pred contains the predicted salaries based on obs of X_test
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

```python
# Visualise the training set results
plt.scatter(X_train , y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue')
plt.title('Salary vs Experience(Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show
```

```python
# Visualise the test set results
plt.scatter(X_test , y_test,color='red')
plt.plot(X_train,regressor.predict(X_train),color='blue') #Predictions of training set and testing set is the same regression line. 
plt.title('Salary vs Experience(Testing Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show
```
